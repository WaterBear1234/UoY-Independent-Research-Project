```python
# Table 36: W-statistics and P-value for the best-case time and cost performance of ACO.
import pandas as pd
from scipy.stats import shapiro

# Define the data
data = {
    'Cost': [20030.68, 56989.57, 951731.45, 587015.15, 1984540.47, 221239.43, 7117374.39, 5670374.32, 18128690.01],
    'Time': [0.57, 1.88, 1.57, 1.82, 1.74, 10.73, 12.41, 45.05, 145.95]
}

# Create a DataFrame
df = pd.DataFrame(data)


# Function to perform the Shapiro-Wilk test
def shapiro_test(dataframe):
    results = {}
    for column in dataframe.columns:
        stat, p_value = shapiro(dataframe[column])
        results[column] = {'W-statistic': stat, 'p-value': p_value}
    return results

# Perform the Shapiro-Wilk test
shapiro_results = shapiro_test(df)

# Print the results and determine the appropriate test
for column, result in shapiro_results.items():
    print(f"Column: {column}")
    print(f"ACO W-statistic: {result['W-statistic']}, ACO p-value: {result['p-value']}")
    if result['p-value'] > 0.05:
        print(f"Use ANOVA for {column}\n")
    else:
        print(f"Use Mann-Whitney U test for {column}\n")
```
