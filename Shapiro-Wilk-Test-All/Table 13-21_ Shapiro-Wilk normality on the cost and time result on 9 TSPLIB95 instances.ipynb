```python
# Table 13: Shapiro-Wilk normality on the cost and time result on the kroD100.tsp instance of TSPLIB95.
import pandas as pd
from scipy.stats import shapiro

# Define the data
data = {
    'Cost': [1484660.057897728, 1442677.7406094524, 1433045.9324615097, 1397849.345850042, 1442224.6100796172, 1449408.816266233, 1360933.9760535618, 1404554.8106816416, 1426302.7495149265, 1420246.1458948313],
    'Time': [0.1233069101969401, 0.12593348821004233, 0.5328218142191569, 0.5270870526631674, 0.31752069791158044, 0.2073546250661214, 0.660637617111206, 2.3483526706695557, 0.7320070266723633, 0.2887304623921712]
}

# Create a DataFrame
df = pd.DataFrame(data)


# Function to perform the Shapiro-Wilk test
def shapiro_test(dataframe):
    results = {}
    for column in dataframe.columns:
        stat, p_value = shapiro(dataframe[column])
        results[column] = {'W-statistic': stat, 'p-value': p_value}
    return results

# Perform the Shapiro-Wilk test
shapiro_results = shapiro_test(df)

# Print the results and determine the appropriate test
for column, result in shapiro_results.items():
    print(f"Column: {column}")
    print(f"W-statistic: {result['W-statistic']}, p-value: {result['p-value']}")
    if result['p-value'] > 0.05:
        print(f"Use ANOVA for {column}\n")
    else:
        print(f"Use Mann-Whitney U test for {column}\n")
```
